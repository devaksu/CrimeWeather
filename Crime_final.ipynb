```python
import pandas as pd
```

### 1. Scripts for reading crime data .csv files and concate into one dataframe: crime_sf

```python
rikos1 = pd.read_csv("7-9_22_SF Crime Data Export â€” 2023-03-10 â€” 05.12.40 PM.csv", sep =',')
rikos2 = pd.read_csv("10-12_22_SF Crime Data Export â€” 2023-03-10 â€” 05.10.49 PM.csv", sep =',')
rikos3 = pd.read_csv("1-3_23_SF Crime Data Export â€” 2023-03-10 â€” 05.08.38 PM.csv", sep =',')

crime_sf = pd.concat([rikos1, rikos2, rikos3], ignore_index=True)
crime_sf
```

### 2. Renaming columns, dropping non-relevant columns , reordering the columns

```python
# NimetÃ¤Ã¤n uudelleen sarakkeet
crime_sf = crime_sf.rename(columns={'incident_datetime': 'datetime', 'analysis_neighborhood': 'neighborhood', 'incident_category' : 'incidentCategory', 'incident_description' : 'incidentDescription'})
# Poistetaan turhat sarakkeet
crime_sf.drop(crime_sf.columns[[0, 4, 5, 6, 8]], axis=1, inplace=True)
# JÃ¤rjestetÃ¤Ã¤n sarakkeet
crime_sf = crime_sf.reindex(columns=['incidentCategory', 'incidentDescription', 'neighborhood', 'datetime'])
crime_sf
```

## 3. Transform datetime column into datetime-object
###       Add column for the hour of the incident

```python

# Muunnetaan sarake 'datetime' datetime muotoiseksi objektiksi 
crime_sf['datetime'] = pd.to_datetime(crime_sf['datetime'])

# #LisÃ¤tÃ¤Ã¤n vuorokauden tunti (muodossa 00, 01, 02 jne.) jolloin rikos tapahtunut
crime_sf['hour'] = crime_sf['datetime'].dt.strftime('%H')
crime_sf
```

###  4. Add datekey column (weather) in format of YYYYMMDDHH

```python
crime_sf['weather'] = (crime_sf['datetime'].dt.strftime('%Y%m%d') + crime_sf['hour']).astype(int)
crime_sf
```

###  5. Replace Nan-values in 'incidentCategory' column with incidentDescriptions

```python
crime_sf['incidentCategory'] = crime_sf['incidentCategory'].fillna(crime_sf['incidentDescription'])
```
